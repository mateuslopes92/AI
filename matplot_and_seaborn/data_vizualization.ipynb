```python
import pandas as pd
from matplotlib import pyplot as plt

df_sales = pd.read_excel("linechart.xlsx");
df_sales.head()
```

```python
# using chart to show the data
plt.figure(figsize=(12,4))  # to make figure fill the whole space

plt.plot(df_sales["Quarter"], df_sales["Fridge"], color="blue", label="Fridge")
plt.plot(df_sales["Quarter"], df_sales["Dishwasher"], color="red", label="Dishwasher")
plt.plot(df_sales["Quarter"], df_sales["Washing Machine"], color="green", label="Washing Machine")

plt.title("Product Sales")
plt.ylabel("Revenue (mln $)")
plt.xlabel("Financial Quarter")

# to show the labels for each line of data
plt.legend()

plt.show()
```

```python
# grouping the data 
total_sales = df_sales[["Fridge", "Dishwasher", "Washing Machine"]].sum()
total_sales
```

```python
# using pie chart
plt.pie(total_sales, labels=total_sales.index, autopct="%1.2f%%", explode=(0.05, 0, 0), shadow=True)
plt.show()
```
