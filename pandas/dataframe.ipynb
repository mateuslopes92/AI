```python
import pandas as pd

dataframe = pd.read_csv("movies.csv")
dataframe.head(3)
```

```python
dataframe.columns
```

```python
# finding all unique values
dataframe.industry.unique()
```

```python
# to count column values
dataframe.industry.value_counts()
```

```python
# to see more than one column like a select (column filtering)
df_new = dataframe[["title", "imdb_rating", "industry"]]
df_new
```

```python
# filtering rows
dataframe[dataframe.release_year>2000]
```

```python
# filtering row with more than one condition
dataframe[(dataframe.release_year>2000) & (dataframe.release_year<=2010)]
```

```python
# another row filtering could be
dataframe[dataframe.studio=='Marvel Studios']
```

```python
# dataframes has describe method which provide statistics from all numerical columns
dataframe.describe()
```

```python
# another function that provide columns statistics is info
dataframe.info()
```

```python
# to get the movie with best rating
dataframe[dataframe.imdb_rating==dataframe.imdb_rating.max()]
```

```python
# to get the movie with best rating and with worst rating
dataframe[(dataframe.imdb_rating==dataframe.imdb_rating.max()) | (dataframe.imdb_rating==dataframe.imdb_rating.min())]
```

```python
# creating columns
dataframe["age"] = dataframe["release_year"].apply(lambda x: 2024 - x)
dataframe.head(4)
```

```python
# using two columns to create a new one where axis 1 mean row values
dataframe["profit"] = dataframe.apply(lambda x: x["revenue"] - x["budget"], axis=1)
dataframe.head(4)
```

```python
# to set an index you can use set_index and implace True will modify original dataframe
dataframe = pd.read_csv("movies.csv")
dataframe.set_index("title", inplace=True)
```

```python
# to print row data
dataframe.loc["Pather Panchali"]
```

```python
# to print row data in more than one value
dataframe.loc[["Pather Panchali", "Doctor Strange in the Multiverse of Madness"]]
```

```python
# to use integer base index
dataframe.iloc[0]
```

```python
# reseting the index
dataframe.reset_index(inplace=True)
dataframe.head(4)
```
