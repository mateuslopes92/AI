```python
import pandas as pd

df = pd.read_csv("weather_data2.csv")
df
```

```python
# some companies or person use -99999 as NA values
# we will treat this data using replace and other technics 
import numpy as np 

df.replace([-99999, -88888], np.nan)
# or
# df.replace(to_replace=[-99999, -88888], value=np.nan)
```

```python
# we can specify the replace by column 
df.replace({
  'temperature': -99999,
  'windspeed': [-99999, -88888],
  'event': 'no event'
}, np.nan)
```

```python
# another option is use the value want to replace
# key old value, value is new value
df.replace({
  -99999: np.nan,
  -88888: np.nan,
  'no event': 'Sunny'
})
```
